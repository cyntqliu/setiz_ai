```python
from datasets import load_dataset

# get streaming dataset to save space
en_mc4 = load_dataset("mc4", "en", split="validation", streaming=True)
```

```python
# initialize the model
from accelerate import init_empty_weights
from transformers import AutoModelForCausalLM

checkpoint = 'mosaicml/mpt-7b'
model = AutoModelForCausalLM.from_pretrained(checkpoint)
```

```python
from accelerate import load_checkpoint_and_dispatch

model.tie_weights()
model = load_checkpoint_and_dispatch(
    model, 'mpt-7b', device_map="auto",
)
```

```python
# some simple data utility functions
from transformers import AutoTokenizer
import torch

tokenizer = AutoTokenizer.from_pretrained(checkpoint)
data_key = "text"
max_length = 256 # full 2048 MSL is too long for generation

def tokenize_function(example):
  tokenized = tokenizer(example[data_key], truncation=True, return_tensors='pt')
  example['input_ids'] = tokenized['input_ids']
  example['attention_mask'] = tokenized['attention_mask']
  return example

def split_example(example):
  # split the tokens in the tokenized example in half
  # return a dataset example with the first half of tokens,
  # and just the second half tokens
  # cap at max_length due to compute restrictions
  input_ids_len = min(max_length, len(example["input_ids"][0]))
  first_half, second_half = [example["input_ids"][0][:input_ids_len//2]], example["input_ids"][0][input_ids_len//2:input_ids_len]

  first_half_ex = {k: v for k, v in example.items() if not(k in ["text", "url", "timestamp"])}
  first_half_ex["input_ids"] = first_half
  # print (first_half_ex)
  # print (second_half)
  return first_half_ex, second_half
```

```python
# sample the dataset - if we are not loading a dataset from file (see 2 cells later)
from datasets import Dataset
import numpy as np
import random

dataset_len = 3083850
num_examples = 1000
sampled_example_ids = random.sample(list(np.arange(dataset_len)), num_examples)
print (f"sampled_example_ids: {sampled_example_ids}")
sampled_examples = []
for idx, ex in enumerate(en_mc4):
  if idx in sampled_example_ids:
    sampled_examples.append(ex)
```

```python
# tokenize the dataset

assert len(sampled_examples) == num_examples
def gen():
  for ex in sampled_examples:
    yield ex

sampled_dataset = Dataset.from_generator(gen)
tokenized_dataset = sampled_dataset.map(tokenize_function)
```

```python
model.to('cuda')
```

```python
# generate continuations for the first halves of sampled examples
from tqdm import tqdm

generations = []
ground_truth_second_halves = []

# turns out 1000 is too much for my resources
for idx, ex in tqdm(enumerate(tokenized_dataset)):
  first_half_ex, second_half = split_example(ex)
  length = len(second_half)
  generation = model.generate(
      torch.tensor(first_half_ex['input_ids']).to(model.device), # not sure why return_tensors is failing
      max_new_tokens=length,
      min_new_tokens=length, # fix num tokens so we get same number of tokens in GT and generations
  ).cpu()

  generations.append(generation[0][-length:]) # do not save the first half
  ground_truth_second_halves.append(second_half)
  if idx >= 99:
    break
```

```python
gen_length = 0
for g in generations:
  gen_length += g.shape[0]
gen_length
```

```python
gt_length = 0
for g in ground_truth_second_halves:
  gt_length += len(g)
assert gt_length == gen_length
```

```python
np.save("ground_truth.npy", ground_truth_second_halves)
np.save("generations.npy", generations)
```

```python
# get relevant token frequencies, sort ground truth by descending frequency
vocab_size = 50432
gt_frequencies = {token_id : 0 for token_id in range(vocab_size)}
for second_half in ground_truth_second_halves:
  for token_id in second_half:
    gt_frequencies[token_id] += 1

generation_frequencies = {token_id : 0 for token_id in range(vocab_size)}
for generation in generations:
  for token_id in generation:
    generation_frequencies[token_id.item()] += 1

sorted_gt_tokens = [token_id for token_id in range(vocab_size)]
sorted_gt_tokens.sort(key=lambda x: gt_frequencies[x], reverse=True)
```

```python
# plot frequency discrepancies for tokens based on their ground truth frequency
import matplotlib.pyplot as plt

# note: sample of data is small enough such that there are no tokens outside of the top 10%
# this means the data is not representative of the set. An actual analysis of this phenomenon
# would involve more resources and a much larger dataset.
x_splits = ["top 10", "top 50", "other top 10%"]
idx_of_x_splits = [10, 50, round(vocab_size/10)]

relative_freqs_in_gen = []
start_idx = 0
for end_idx in idx_of_x_splits:
  gt_count = sum([gt_frequencies[token_id] for token_id in sorted_gt_tokens[start_idx:end_idx]])
  gen_count = sum(
      [generation_frequencies[token_id] for token_id in sorted_gt_tokens[start_idx:end_idx]]
  )
  diff = gen_count - gt_count
  percent_diff = (diff / gt_count) * 100
  relative_freqs_in_gen.append(percent_diff)
  start_idx = end_idx

fig, ax = plt.subplots()
ax.bar(x_splits, relative_freqs_in_gen)
ax.set_ylabel("% difference in freq")
ax.set_title("Difference between Generated and Ground Truth Token Frequency")
plt.setp(ax.get_xticklabels(), fontsize=10, rotation='vertical')

plt.show()
```
